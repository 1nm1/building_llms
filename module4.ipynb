# Approach
My approach is to do something similar to what we’ve done in class in that we generate embeddings for the query and the isolated text and then compute a cosine similarity between them, however, I'd like to add a keyword creation transformer and then pass it to a aspect-based sentiment analysis transformer to more appropriate list of hotels

```python
aste_classifier = ASTE.AspectSentimentTripletExtractor("english")
```

```python
nlp = spacy.load("en_core_web_sm")

def find_keywords_spacy(text):
    doc = nlp(text)
    nouns = [token.text for token in doc if token.pos_ == "NOUN"]
    return nouns

def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    
    # Remove newlines
    text = text.replace('\n', ' ').replace('\r', '')
    
    # Replace multiple spaces with a single space
    text = re.sub(' +', ' ', text)
    
    return text.strip()


def return_object_aspects(reviews):
    result_dict = {}
    polarity_to_value = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
    for review in reviews:
        aspects = aste_classifier.predict(clean_text(review))['Triplets']
        aspect_dict = {}
        print(f"\n\n{aspects}")

        # Iterate through the list of dictionaries
        for item in aspects:
            print(item)
            if isinstance(item, dict):
                aspect = item['Aspect']
                polarity = polarity_to_value[item['Polarity']]
                
                # If the aspect is not in the accumulator, add it
                if aspect not in aspect_dict:
                    aspect_dict[aspect] = 0

                # Accumulate the polarity value for the current aspect
                aspect_dict[aspect] += polarity
        # there is an issue here where if you update something and the
        # key is already present, it'll override that value
        result_dict.update(aspect_dict)
    
    return result_dict


def query_similarity(df, query, n=5, pprint=True):
    print("=== Starting search ===")

    print(f"Query: {query}")
    query_keywords = find_keywords_spacy(query)
    print(f"Query keywords: {query_keywords}")

    if 'embedding' not in df.columns.to_list():
      print('ERROR: dataframe does not contain embeddings')
      return None

    print("Generating embeddings and cosine similarity")
    query_embedding = embedder.encode(query,show_progress_bar=True)
    df["similarity"] = df.embedding.apply(lambda x: cosine_similarity(x, query_embedding.reshape(768,-1)))

    results = (df.sort_values("similarity", ascending=False).head(n))

    resultlist = []
    hotel_list = []

    # Generate list based on cosine similarity
    for r in results.index:
        if results.name[r] not in hotel_list:
            hotel_df = results.loc[results.name == results.name[r]]

            reviews = [hotel_df.review[s] for s in hotel_df.index]

            aspects = return_object_aspects(reviews)
 
            resultlist.append(
            {
              "name":results.name[r],
              "score": hotel_df.similarity[r][0],
              "rating": hotel_df.rating.max(),
              "price": hotel_df.priceLevel[r],
              "reviews": reviews,
              "review_aspects": aspects
            })
            hotel_list.append(results.name[r])

    # Further refine list by looking at keywords and aspect based sentient analysis
    
    result_df = pd.DataFrame(resultlist)
    normalized_df = pd.json_normalize(result_df['review_aspects'])
    normalized_df.columns = ['_' + col for col in normalized_df.columns]
    final_df = pd.concat([result_df, normalized_df], axis=1)

    filtered_results = {}
    for word in query_keywords:
        print(word)
        print(final_df.columns.to_list())
        for col in final_df.columns.to_list():
            if f"_{word.lower()}" == col:
                filtered_results.update(
                    {word: final_df.sort_values(col, ascending=False).head(n)}
                )

    return final_df, filtered_results
```

```python
query = 'Hotel with great pool and shops nearby'
results, filtered_results = query_similarity(df_w_embeddings,query,20)
```
